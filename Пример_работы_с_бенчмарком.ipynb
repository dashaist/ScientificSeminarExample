```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel, T5Tokenizer, T5ForConditionalGeneration
import numpy as np
import matplotlib.pyplot as plt
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

# Загрузка моделей
gpt2_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
gpt2_model = GPT2LMHeadModel.from_pretrained("gpt2")
gpt2_tokenizer.pad_token = gpt2_tokenizer.eos_token  # Фикс для генерации

t5_tokenizer = T5Tokenizer.from_pretrained("t5-small")
t5_model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Функция генерации через GPT-2
def generate_gpt2(question, max_length=100):
    inputs = gpt2_tokenizer.encode(question, return_tensors="pt")
    outputs = gpt2_model.generate(
        inputs,
        max_length=max_length,
        num_return_sequences=1,
        pad_token_id=gpt2_tokenizer.eos_token_id,
        temperature=0.7,  # Снижаем случайность
        top_p=0.9,        # Фильтрация ядра
    )
    return gpt2_tokenizer.decode(outputs[0], skip_special_tokens=True)

# Функция генерации через T5
def generate_t5(question, context, max_length=100):
    input_text = f"question: {question} context: {context}"
    inputs = t5_tokenizer.encode(input_text, return_tensors="pt")
    outputs = t5_model.generate(
        inputs,
        max_length=max_length,
        num_return_sequences=1,
        temperature=0.5,
        top_p=0.85,
    )
    return t5_tokenizer.decode(outputs[0], skip_special_tokens=True)

# Метрики с улучшенным BLEU
def evaluate_metrics(reference, candidate):
    smooth = SmoothingFunction().method1
    bleu = sentence_bleu(
        [reference.split()],
        candidate.split(),
        smoothing_function=smooth
    )

    # Косинусное сходство
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
    emb_ref = model.encode([reference])
    emb_cand = model.encode([candidate])
    cosine = cosine_similarity(emb_ref, emb_cand)[0][0]

    # ROUGE-L (упрощенный)
    rouge = len(set(reference.split()) & set(candidate.split())) / len(set(reference.split()))

    return {"BLEU": bleu, "Cosine": cosine, "ROUGE-L": rouge}

# Тестовые данные
test_data = [
    {
        "question": "What are the symptoms of COVID-19?",
        "context": "COVID-19 symptoms include fever, cough, fatigue, and loss of taste or smell.",
        "answer": "The main symptoms are fever, cough, fatigue, and loss of taste or smell."
    },
    {
        "question": "How does photosynthesis work?",
        "context": "Photosynthesis converts sunlight into chemical energy in plants using chlorophyll.",
        "answer": "It is the process by which plants convert sunlight into energy using chlorophyll."
    }
]

# Оценка
results = {"GPT-2": [], "T5": []}

for item in test_data:
    question = item["question"]
    context = item["context"]
    answer = item["answer"]

    # Генерация
    gpt2_response = generate_gpt2(question)
    t5_response = generate_t5(question, context)

    # Оценка
    results["GPT-2"].append(evaluate_metrics(answer, gpt2_response))
    results["T5"].append(evaluate_metrics(answer, t5_response))

    print(f"Вопрос: {question}")
    print(f"GPT-2: {gpt2_response}")
    print(f"T5: {t5_response}")
    print("---")

# Визуализация
metrics = ["BLEU", "Cosine", "ROUGE-L"]
models = ["GPT-2", "T5"]

avg_scores = {
    model: {metric: np.mean([r[metric] for r in results[model]]) for metric in metrics}
    for model in models
}

plt.figure(figsize=(10, 5))
for i, metric in enumerate(metrics):
    plt.subplot(1, 3, i+1)
    plt.bar(models, [avg_scores[model][metric] for model in models])
    plt.title(metric)

plt.tight_layout()
plt.show()
```
